```python
height=int(input())
for i in range(height):
    for j in reversed(range(height)):
        if j > i:
            print(' ', end='')
        else:
            print('*', end='')  #삼각형 반쪽
    for j in range(height):
        if j < i:
            print('*', end='')
    print()
```

# 20. FizzBuzz 문제 

```python
for i in range(1,101):  #끝을 101로 지정!!!!
    if i%15==0:
        print("FizzBuzz")
    elif i%5==0:
        print("Buzz")
    elif i%3==0:
        print("Fizz")
    else:
        print(i)
```

```python
for i in range(1,101):
    print("Fizz" * (i%3 ==0) + 'Buzz'* (i % 5 ==0) or i)
#코드 골프 방식: 문자 수를 최대한 줄여서 작성 but 가독성 떨어져 실무에선 비추
```

```python
for i in range(1,101):
    if i%2==0 and i%11==0:
        print('FizzBuzz')
    elif i%2==0:
        print('Fizz')
    elif i%11==0:
        print('Buzz')
    else:
        print(i)
    
```
