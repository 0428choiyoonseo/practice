 - self.메서드( ): 메서드 안에서 메서드 호출 <br>
    -self. 없이 메서드 이름만 사용시 클래스 바깥쪽에 있는 함수를 호출한다는 뜻 <br>
 - isinstance( 인스턴스, 함수 ) :  현재 인스턴스가 특정 클래스의 인스턴스인지 확인


- 속성(attribute) 만들기: __init_ 메서드 안에  self.속성에 값 할당

```python
class Person:
    def __init__(self):
        self.hello = '안녕하세요.' #Person 클래스의 __init__ 메서드에서 self.hello에 '안녕하세요.' 인사말을 넣음
    def greeting(self):
        print (self.hello)

james = Person()  #james 객체 생성
james.greeting()  #메서드 호출
```

- __init__ 메서드 : 클래스에 ( )(괄호)를 붙여서 인스턴스를 만들 때 호출되는 특별한 메서드 <br>

    : 클래스로 객체를 만들 때 자동으로 실행되는 초기화 함수
    -initiallize 이름 그대로 인스턴스(객체) 초기화 <br>
-스페셜 메서드(special method) / 매직 메서드(magic method) : 앞뒤로 __가 붙은 메서드. 파이썬이 자동으로 호출해주는 메서드

- self : 인스턴스 자기 자신 <br>
    : 메서드를 호출한 객체 자기 자신을 가리키는 매개변수. 객체의 속성이나 다른 메서드에 접근하기 위해 사용 

- 인스턴스 속성 :  인스턴스를 통해 접근하는 속성 <br>
    인스턴스.속성 : 클래스 바깥에서 속성에 접근

```python
class Person:
    def __init__(self, name, age, address):     
        self.hello = '안녕하세요.'  
        self.name = name
        self.age = age
        self. address = address
        
    def greeting(self):     #이 때 __init__자동을 실행됨
        print('{0} 저는 {1}입니다'.format(self.hello, self.name))
        
maria = Person ('마리아' , 20, '서울시 서초구 반포동')
maria.greeting( ) #안녕하세요. 저는 마리아입니다.


print('이름:', maria.name)       # 마리아
print('나이:', maria.age)        # 20
print('주소:', maria.address)    # 서울시 서초구 반포동
```

__slots__ = ['속성이름', '속성이름2'] <br>
:허용할 속성 이름을 리스트로 넣어줌. 특정 속성만 허용할 때 사용. 속성 이름은 반드시 문자열.

- 비공개 속성(private attribute) : 클래스 바깥에서는 접근할 수 없고 클래스 안에서만 사용 가능한 속성\
    형식: __속성명
- 공개 속성(public attribute) : 클래스 바깥에서 접근할 수 있는 속성

*메서드도 __(밑줄 두 개)로 시작하면 클래스 안에서만 호출할 수 있는 비공개 메서드가 됨

```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet
    def pay(self, amount):
        self.__wallet-= amount #비공개 속성은 클래스 안의 메서드에서만 접근 가능
        print('이제 {0}원 남았네요.'.format(self.__wallet))
maria = ('마리아', 20, '서울시 서초구 반포동', 10000)
maria.pay(3000)
```

```python
class Knight:
    def __init__(self, health, mana, armor):
        self.health=health
        self.mana=mana
        self.armor=armor
    def slash(self):
        print('베기')

x = Knight(health=542.4, mana=210.3, armor=38)
print(x.health, x.mana, x.armor)
x.slash()
```

```python
class Annie:
    def __init__(self, health, mana, ability_power):
        self.health=health
        self.mana=mana
        self.ability_power=ability_power
    def tibbers(self):
        print('티버:  피해량 ', self.ability_power*0.65+400)
                                            
        
health, mana, ability_power = map(float, input().split())
 
x = Annie(health=health, mana=mana, ability_power=ability_power)
x.tibbers()        
```
