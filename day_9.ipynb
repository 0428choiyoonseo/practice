-  클래스 속성 사용하기 <br>
    class 클래스이름: <br>
        속성 = 값 <br>

```python
class Person:
    bag = []        #bag 속성 넣기
    
    def put_bag(self, stuff):   
        self.bag.append(stuff)
        
james = Person( )
james.put_bag('책')
    
maria = Person( )
maria. put_bag('열쇠')
    
print(james.bag)
print(maria.bag) 
```

-> 클래스 속성은 모든 인스턴스에서 공유하기에 넣었던 물건이 합쳐져서 출력됨. <br>
    put_bag 메서드에서 클래스 속성 bag에 접근할 때 self를 사용했지만    <br>
    -> self는 현재 인스턴스를 뜻하므로 클래스 속성을 지칭하기에는 조금 모호 <br>

클래스 속성에 접근할 때는 클래스 이름으로 접근 <br>
--> 클래스.속성 

```python
class Person:
    def __init__(self):
        self.bag = []        #bag을 인스턴스 속성으로 만들면 가방을 여럿이서 공유 x
                                         #인스턴스 속성은 인스턴스별로 독립되어 있으며 서로 영향 x
    def put_bag(self, stuff):       
        self.bag.append(stuff)          
 
james = Person()
james.put_bag('책')
 
maria = Person()
maria.put_bag('열쇠')
 
print(james.bag)
print(maria.bag)
```

클래스의 독스트링은 클래스.__ doc__ 형식  <br>
메서드의 독스트링은 클래스.메서드.__ doc__ 또는 인스턴스.메서드.__ doc__ 형식 <br>


# 정적 메서드
    -메서드 위에 @staticmethod 붙임
    -매개변수에 self 지정 x  -> 인스턴스 속성엔 접근 못함
        -->  그래서 인스턴스 속성/메서드가 필요 없을 때 사용

- 데코레이터 :  앞에 @가 붙은 것. 메서드에 추가 기능을 구현
- 클래스.메서드() : 정적메서드 호출할 때 클래스에서 바로 메서드 호출

```python
class Calc:
    @staticmethod
    def add(a,b):
        print( a+b )
        
    @staticmethod
    def mul(a,b):
        print(a*b)
Calc.add(10. 20)    #클래스에서 바로 메서드 호출
Calc.mul(10, 20)
```

-정적 메서드는 실행이 외부 상태에 영향을 끼치지 않는 순수 함수(pure function)를 만들 때 사용 <br>
순수함수 : 부수 효과(side effect)가 없고 입력 값이 같으면 언제나 같은 출력 값을 반환하는 함수

# 클래스 메서드
    -메서드 위에 @classmethod 붙임
    -첫 번째 매개변수에 cls 지정해야함
    -클래스 속성, 클래스 메서드에 접근해야 할 때 사용

```python
class Person:
    count = 0   #클래스 속성
    
    def __init__(self):
        Person.count += 1  #인스턴스가 만들어질 때 클래스 속성 count에 1 더함
        
    @classmethod
    def print_count(cls):
        print('{0}명 생성되었습니다.'.format(cls.count))  #cls로 클래스 속성에 접근
        
james = Person()
maria = Person()

Person.print_count()    #2명 생성되었습니다.
```

cls는 클래스 이므로 cls()는 Person()과 같음

```python
class Date :
    
    @staticmethod 
    def is_date_valid(date_string):         #첫 번째 매개 변수로 date_string(날짜 문자열)을 지정 
        year, month, day = map(int, date_string.split('-'))
        return month <= 12 and day <= 31
    
if Date.is_date_valid('2000-10-31'):
    print('올바른 날짜 형식입니다.')
else:
    print('잘못된 날짜 형식입니다.')
```

```python
class Time:
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second
    
    @staticmethod               #문자열이 올바른 시간인지 검사만 하면 되고, 클래스에 접근할 필요 x. -> 정적 메서드로 제작
    def is_time_valid(time_string):
        hour, minute, second = map(int, time_string.split(':'))
        return hour <= 24 and minute <= 59 and second <= 60
        

    @classmethod
    def from_string(cls,time_string):
        hour, minute, second = map(int, time_string.split(':'))
        return cls(hour, minute, second) #클래스에 변수 넣어서 인스턴스 만든 후 바깥에서 사용할 수 있도록 return
        
        
time_string = input()
 
if Time.is_time_valid(time_string):
    t = Time.from_string(time_string)
    print(t.hour, t.minute, t.second)
else:
    print('잘못된 시간 형식입니다.')
```

- 파생 클래스(derivedc class) : 상속을 받아 새롭게 만드는 클래스 ( =자식 클래스(child class), 서브 클래스(subclass))


class 파생클래스이름(기반클래스이름):

```python
class Person:
    def greeting(self):
        print('안녕하세요.')
class Student(Person):
    def study(self):
        print('공부하기')
        
james = Student()       
james.greeting()
james.study()
        
```
