- 세트의 합집합  <br>
세트1 | 세트2 <br>
set.union(세트1, 세트2)
- 세트의 교집합 <br>
    세트1 & 세트2 <br>
    set.intersection(세트1, 세트2) <br>
- 세트 차집합 <br>
    세트1-세트2 <br>
    set.difference(세트1, 세트2) <br>
- 세트 대칭차집합 :  두 집합 중 겹치지 않는 요소만 포함 <br>
    세트1 ^ 세트2 <br>
    set.symmetric_difference(a,b) <br>

```python
a = {1, 2, 3, 4}
a|={5}  # = a.update({5})
a
```

```python
 a = {1, 2, 3, 4}
 a &= {0, 1, 2, 3, 4} # = a.intersection_update({0, 1, 2, 3, 4})
 a
```

```python
a = {1, 2, 3, 4}
a -= {3} # = a.difference_update({3})
a
```

```python
 a = {1, 2, 3, 4}
a ^= {3, 4, 5, 6} # = a.symmetric_difference_update({3, 4, 5, 6})
a
```

- 자동으로 파일 객체 닫기 <br>

```python
with open('hello txt', 'r') as file:
    s = file.read()
    print(s)
```

```python
with open ('hello txt', 'w')as file: #hello.txt 파일을 쓰기 모드(w)로 열기
    for i in range(3):
        file.write('Hrllo,world! {0}\n'.format(i))
```

- 파일객체.writelines(문자열리스트)  
    각 문자열 끝에 \n 붙여주지 않으면 한줄로 붙어서 저장됨

```python
lines = ['안녕하세요.\n', '파이썬\n', '코딩 도장입니다.\n']
with open('hello txt','w') as file:
    file.writelines(lines)
```

- 변수 = 파일객체.readline() : 파일 내용을 한 줄씩 순차적으로 읽음 <br>
    readline 으로 파일 읽을 땐 파일에 문자열이 몇개 있는지 모르기에 while 반복문 활용

```python
with open('hello txt','r') as file:
    line = None
    while line !='': #빈 문자열 아닐 때 계속 반복
        line = file.readline( )
        print(line.strip("\n")) #파일에서 읽어온 문자열에서 \n 삭제하여 출력
```

- for line in file : 파일의 내용 한 줄씩 읽기

```python
with open('hello txt','r')as file:   #읽기 모드로 열기
    for line in file:  #for에 파일 객체 지정 하면 파일의 내용 한 줄씩 읽어서 변수에 저장
        print(line.strip('\n')) #읽어온 문자열에서 \n 삭제하여 출력
```

*파일 객체는 이터레이터

- 피클링(pickling) : 파이썬 객체를 파일에 저장하는 과정
- 언피클링(unpickling) : 파일에서 객체를 읽어오는 과정


- pickle.dump로 객체(값)를 저장할 때는 open('james.p', 'wb')와 같이 파일 모드를 'wb'로 지정 <br>
     b는 바이너리(binary)를 뜻하는데, 바이너리 파일은 컴퓨터가 처리하는 파일 형식 <br>
     .txt 파일은 사람이 알아보기 쉽도록 만든 파일 형식이며 텍스트(text) 파일 <br>

```python
import pickle
name = 'james'
age = 17
address = '서울시 서초구 반포동'
scores = {'korean': 90, 'english': 95, 'mathematics': 85, 'science': 82}
with open('james.p','wb') as file: #파일을 바이너리 쓰기 모드(wb)로 열기
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(address, file)
    pickle.dump(scores, file)
 
```

- 언피클링  <br>
 name, age, address, scores 순으로 저장했으므로 가져올 때도 같은 순서로 가져오기

```python
import pickle
 
with open('james.p', 'rb') as file:    # james.p 파일을 바이너리 읽기 모드(rb)로 열기
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)
    print(name)
    print(age)
    print(address)
    print(scores)
```

28. 회문 팔별, N-gram 만들기

- 회문:순서를 거꾸로 읽어도 제대로 읽은 것과 같은 단어와 문장

```python
word=input('단어를 입력')
is_palindrome = True #회문 판볇갓 저장,초기값 True
for i in range (len(word)//2): #0부터 문자열 길이의 절반만큼 반복
    if word[i]!=word[-1-i]: #왼쪽 문자와 오른쪽 문자를 비교해 문자가 다르면
        is_palindrome=False
        break
print(is_palindrome)
```

```python
word=input('단어를 입력')
print(word==word[::-1]) #원래 문자열과 반대로 뒤집은 문자열 비교
```

- N-gram: 믄지얄에서 N개의 연속된 요소 추출하는 방법

```python
text='Hello'
for i in range(len(text)-1):  #2-gram이므로 문자열의 끝에서 한글자 앞까지만 반복. 멘 뒤에선[i+1]이 없으니까
    print(text[i], text[i+1], sep='') #현재 문자오 그 다음 문자 출력
```

```python
text='hello'
two_gram = zip(text,text[1:])
for i in two_gram:    #문자열은 시퀀스객체라 for문 가능
    print(i[0],i[1], sep='')
```
