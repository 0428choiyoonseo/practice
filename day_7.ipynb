- filter : filter에 지정한 함수의 반환값이 True일 때 반복 가능한 객체에서 조건에 맞는 요소만 가져옴 <br>
filter  (함수, 반복가능한객체)

```python
def f(x):
    return x > 5 and x<10
a = [8, 3, 2, 10, 15, 7, 1, 9, 0, 11]
list(filter(f,a))
```

```python
a = [8, 3, 2, 10, 15, 7, 1, 9, 0, 11]
list(filter(lambda x:x>5 and x <10, a))
```

# 33. 클로저 사용

- 전역변수 :함수를 포함하여 스크립트 전체에서 접근할 수 있는 변수

```python
x = 10          # 전역 변수
def foo():
    x = 20      # x는 foo의 지역 변수
    print(x)    # foo의 지역 변수 출력
 
foo()
print(x)        # 전역 변수 출력

# 전역변수 x와 지역변수 x는 이름만 같을 뿐 다른 변수
```

*함수 안에서 전역 변수의 값을 변경하려면 global 키워드를 사용 <br>
global x 하고 변수 설정

*nonlocal : 함수 바깥쪽에 있는 지역 변수의 값을 변경 <br>
    가장 가까운 바깥쪽 함수의 지역변수 사용. 없으면 점점 바깥쪽으로 나감

```python
def A():
    x = 10        # A의 지역 변수 x
    def B():
        nonlocal x    # 현재 함수의 바깥쪽에 있는 지역 변수 사용
        x = 20        # A의 지역 변수 x에 20 할당
 
    B()
    print(x)      # A의 지역 변수 x 출력
 
A()
```

- closure : 함수를 둘러싼 환경(지역 변수, 코드 등)을 계속 유지하다가, 함수를 호출할 때 다시 꺼내서 사용하는 함수 <br>
    지역 변수와 코드를 묶어서 사용하고 싶을 때 활용 <br>
    클로저가 속한 지역 변수는 바깥에서 직접 접근 할 수 없으므로 데이터 숨기고싶을 때 활용
    

```python
def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b    # 함수 바깥쪽에 있는 지역 변수 a, b를 사용하여 계산. 함수 calc가 끝났는데도 c는 calc의 지역 변수 a, b를 사용해서 계산
    return mul_add
c=calc()
print(c(1), c(2), c(3), c(4), c(5))
```

lambda로 클로저 만들기

```python
def calc():
    a = 3
    b = 5
    return lambda x : a*x+b #람다 표현식을 반환
x = calc()
print(c(1), c(2), c(3), c(4), c(5))
```

```python
def calc():
    a= 3
    b =5
    total = 0
    def mul_add(x):
        nonlocal total              #클로저의 지역 변수 변경하고싶으면 nonlocal 사용
        total = total + a*x + b
        print(total)
    return mul_add

c= calc()
c(1)
c(2)
c(3)

```

```python
def countdown(n):       
     i = n+1    #지역변수 i 에 매개변수 n에 1을 더해서 할당
     
     def count():
             nonlocal i 
             i -= 1     
             return i 
     return count   
n = int(input())        
c = countdown(n)  # countdown 함수 호출에 반환값을 c에 저장한 뒤 c 호풀
for i in range(n):
    print(c(), end=' ')
```
